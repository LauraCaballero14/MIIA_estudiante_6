```python
# PARTE 1 MISION 1: Ciclo que automatiza la lectura de los archivos

import os

direccion_archivos='C:/Users/USER/Documents/PERSONAL/Maestria Andes/1-Herramientas Computacionales/HCAD-main/Lab 1/Archivos'

def abrir_archivos_diccionario(ruta_archivos):

    lista_archivos=os.listdir(ruta_archivos) #Convierte los nombres de una ruta especificada en una lista
    
    datos_completos={}
    for archivo in lista_archivos:
        if archivo.endswith('.txt'): #Indica la terminación del archivo 
            ruta=".\\Archivos\\" + archivo
    
            with open(ruta, encoding="utf-8") as informacion:
                llave=archivo[:-4] #Elimina la terminación txt para hacer la llave
                datos=informacion.readlines()
        
                datos_completos[llave]=datos
        else:
            pass

    return datos_completos

diccionario_becas=abrir_archivos_diccionario(direccion_archivos)

diccionario_becas
```

```python
# PARTE 1 MISION 2: Metricas básicas

# función que calcula la media de una lista de datos numéricos

def promedio(l):
    prom=sum(l)/len(l)
    return round(prom,2)

# función que calcula la desviación de una lista de datos numéricos

def desviacion(l):
    c = promedio(l)
    ss = sum((x-c)**2 for x in l)
    var_muestral=ss/(len(l)-1)
    return round(var_muestral**0.5,2)

#Número de registros
print("La cantidad de observaciones para analizar la distribución de becas es " + str(len(diccionario_becas["edad"])))

#Metricas básica de las variables cuantitativas

#Edad
print("La media de las edades es " + str(promedio(diccionario_becas["edad"])))
print("La desviación de las edades es " + str(desviacion(diccionario_becas["edad"])))
print("La edad minima es " + str(min(diccionario_becas["edad"])))
print("La edad máxima es " + str(max(diccionario_becas["edad"])))

#Promedio
print("La media del promedio es " + str(promedio(diccionario_becas["promedio"])))
print("La desviación de los promedios es " + str(desviacion(diccionario_becas["promedio"])))
print("El promedio más bajo es " + str(min(diccionario_becas["promedio"])))
print("El promedio más alto es " + str(max(diccionario_becas["promedio"])))

#Estrato
print("La media del estrato es " + str(promedio(diccionario_becas["estrato"])))
print("La desviación de los estratos es " + str(desviacion(diccionario_becas["estrato"])))
print("El estrato más bajo es " + str(min(diccionario_becas["estrato"])))
print("El estrato más alto es " + str(max(diccionario_becas["estrato"])))

#Variables cualitativas

#Función que cuenta los datos por nivel de cada factor. Ejemplo: Número de registros en el nivel femenino para el factor genero
def conteo_niveles(lista):
    dicc_conteo_niveles={}
    niveles=list(set(lista))
    for i in niveles:
        dicc_conteo_niveles[i]=lista.count(i)
    return dicc_conteo_niveles

niveles_escolaridad=conteo_niveles(diccionario_becas["escolaridad"])
niveles_estado=conteo_niveles(diccionario_becas["estado_civil"])
niveles_genero=conteo_niveles(diccionario_becas["genero"])
niveles_region=conteo_niveles(diccionario_becas["region"])
niveles_estrato=conteo_niveles(diccionario_becas["estrato"])

print(niveles_escolaridad)
print(niveles_estado)
print(niveles_genero)
print(niveles_region)
print(niveles_estrato)

# función que calcula la moda 

def moda(diccionario):
    valor_moda=max(diccionario.values())
    for key,value in diccionario.items():
        if valor_moda==value:
            return key
        
print("La moda en escolaridad es "+ str(moda(niveles_escolaridad)))
print("La moda en estado civil es "+ str(moda(niveles_estado)))
print("La moda en genero es "+ str(moda(niveles_genero)))
print("La moda en region es "+ str(moda(niveles_region)))
print("La moda en estrato es "+ str(moda(niveles_estrato)))



```

**Misión 4**

Desarrollar un algoritmo que permita conocer la población que sería becada según los criterios propuestos por el *primer grupo* mencionado arriba.

***Racional del autor para desarrollar mision 4***

1. Se identifica el número de becas a asignar por nivel en la primera ronda. Es decir se obtiene el 2% de registros de cada estrato. 

2. La cantidad de becas a asignar se divide por la suma de las becas repartidas en esta primera ronda, obteniendo el número  de rondas que se deben realizar para asignar el 2% de cada nivel.

3. Se calcula el residuo entre las becas y las becas asignadas en las rondas, para repartirlas por el orden ascendente de los estratos, obteniendo así el número total de becados por estrato.

4. Se genera una lista de tuplas, donde cada elemento de la lista es una tupla con el ID, la edad, escolaridad, estado civil, estrato, genero, region y promedio.

5. Se ordena esta lista de tuplas por promedio, pero como la tupla es inmutable no se pierde la identificación ni los datos de estos registros.

6. Para cada estrato se rebana la lista de tuplas, tomando la cantidad de becas obtenidas en el punto 3 en la lista organizada por promedio.

7. Se usa lo implementado en la mision 2 para conocer la población becada por este criterio.





```python
#PARTE 1 MISION 4: IDENTIFICAR EL NÚMERO DE BECAS A ASIGNAR EN LA PRIMERA RONDA

import math

becas=int(input("Ingrese el número de becas a asignar "))

#Una lista con el factor estrato y otra con la cantidad de observaciones con

lista_niveles_estrato=list(niveles_estrato.keys())
lista_valores_estrato=list(niveles_estrato.values())


#1. Se identifica el número de becas a asignar por nivel en la primera ronda. 
#Es decir se obtiene el 2% de registros de cada estrato. 

lista_becas_estrato=[]
for i in range(len(lista_valores_estrato)):
    lista_becas_estrato.append(math.ceil(lista_valores_estrato[i]*0.02))
 

#2. La cantidad de becas a asignar se divide por la suma de las becas repartidas en esta primera ronda, 
#obteniendo el número  de rondas que se deben realizar para asignar el 2% de cada nivel.

cantidad_repeticiones=becas//sum(lista_becas_estrato)

#Cantidad de becas por estrato repartiendo exactamente el 2%
lista_becas_estrato_modulo=[]
for i in range(len(lista_becas_estrato)):
    lista_becas_estrato_modulo.append(math.ceil(lista_becas_estrato[i]*cantidad_repeticiones))

#3. Se calcula el residuo entre las becas y las becas asignadas en las rondas, para repartirlas por el orden ascendente 
#de los estratos, obteniendo así el número total de becados por estrato.
    
residuo=becas-sum(lista_becas_estrato_modulo)
print(residuo)

#Cantidad de becas por estrato que faltaron por repartir
cantidad_becas_estrato_residuo=[]
for i in range (len(lista_becas_estrato)):
    if sum(cantidad_becas_estrato_residuo[0:i+1])+lista_becas_estrato[i]<=residuo:
        cantidad_becas_estrato_residuo.append(lista_becas_estrato[i])
    elif sum(cantidad_becas_estrato_residuo[0:i+1])+lista_becas_estrato[i]>residuo:
        cantidad_becas_estrato_residuo.append(residuo-sum(cantidad_becas_estrato_residuo[0:i+1]))
    else:
        cantidad_becas_estrato_residuo.append(0)
        

#Cantidad de becas asignadas por estrato
cantidad_becas_por_estrato_final=[]
for i in range(len(lista_becas_estrato)):
    cantidad_becas_por_estrato_final.append(cantidad_becas_estrato_residuo[i]+lista_becas_estrato_modulo[i])

cantidad_becas_por_estrato_final

```

```python
#PARTE 2 MISION 4: IDENTIFICAR LOS BECADOS

# 4. Se genera una lista de tuplas, donde cada elemento de la lista es una tupla con el ID, 
#la edad, escolaridad, estado civil, estrato, genero, region y promedio.
tuplas_becas=[]
for i in range(len(list(diccionario_becas["ID"]))):
    tuplas_becas.append((diccionario_becas["ID"][i],diccionario_becas["edad"][i],
                        diccionario_becas["escolaridad"][i],diccionario_becas["estado_civil"][i],
                        diccionario_becas["estrato"][i],diccionario_becas["genero"][i],
                       diccionario_becas["region"][i],diccionario_becas["promedio"][i]))
tuplas_becas.sort(key = lambda x: x[-1], reverse=True)

#Se genera una lista de tuplas con el estrato y el numero de observaciones por estrato
lista_tuplas_estrato=[]
for i in range(len(lista_niveles_estrato)):
    lista_tuplas_estrato.append((lista_niveles_estrato[i],lista_valores_estrato[i]))

#6. Para cada estrato se rebana la lista de tuplas, 
#tomando la cantidad de becas obtenidas en el punto 3 en la lista organizada por promedio.

#Definir función para rebanar de la lista de acuerdo a la cantidad de becas por estrato

def rebanado_estrato(listabecas, listaniveles,indicenivel,listabecados):
    listarebanada=[]
    for i in range(len(listabecas)):
        if listabecas[i][4]==listaniveles[indicenivel][0]:
            listarebanada.append(listabecas[i])
            n=int(listabecados[indicenivel])
    return listarebanada[:n]

#Lista de tuplas con becados
becas_estrato=[]
for i in range(len(lista_tuplas_estrato)):
    becas_estrato.extend(rebanado_estrato(tuplas_becas,lista_tuplas_estrato,i,cantidad_becas_por_estrato_final))

print("La cantidad de registros becados son "+ str(len(becas_estrato)))
becas_estrato
```

```python
# PARTE 1 MISION 7 PARAMETROS A INGRESAR Y ORGANIZACION DE DATOS

#Rango Edades

lim_inf_edad=int(input("Ingrese la edad minima para ser candidato a tener una beca: "))
lim_sup_edad=int(input("Ingrese la edad maxima para ser candidato a tener una beca: "))  

#Organizar la lista de tuplas por promedio
tuplas_becas.sort(key=lambda x:x[-1],reverse=True)

#Filtrar por el rango de edad
tuplas_becas_edad=[]
for i in range(len(tuplas_becas)):
    if tuplas_becas[i][1]>=lim_inf_edad and tuplas_becas[i][1]<=lim_sup_edad:
        tuplas_becas_edad.append(tuplas_becas[i])
        

#Elegir criterio
criterio_cualitativo=int(input("Ingrese el dígito relacionado con el criterio de asignación de becas: "))

#Indicador de la posición en la lista de tuplas del criterio cuantitativo elegido
if criterio_cualitativo==1:
    posicion=5
elif criterio_cualitativo==2:
    posicion=4
elif criterio_cualitativo==3:
    posicion=6

#Generar una lista unicamente con el criterio elegido
lista_criterio=[]
for i in range(len(tuplas_becas_edad)):
    lista_criterio.append(tuplas_becas_edad[i][posicion])

#Hacer un diccionario que cuente los niveles por el criterio elegido    
niveles_criterio=conteo_niveles(lista_criterio)

#Hacer una lista ordenada con estas tuplas por cantidad de registros 
lista_tuplas_criterio_orden=[]
for i in range(len(list(niveles_criterio.keys()))):
    lista_tuplas_criterio_orden.append((list(niveles_criterio.keys())[i],list(niveles_criterio.values())[i]))
lista_tuplas_criterio_orden.sort(key = lambda x: x[-1], reverse=True)

#Ingresar las proporciones por cada nivel, los criterios elegidos (estrato y region) contienen 5 niveles mientras que genero 4.
proporciones=[]
cantidad_niveles=len(lista_tuplas_criterio_orden)

for i in range(0,cantidad_niveles-1):
    proporciones.append(float(input("Proporcion para el nivel (número del 0 al 100) " +str(lista_tuplas_criterio_orden[i][0])))/100)

prop5=1-sum(proporciones)
proporciones.append(prop5)

if proporciones[-1]<0:
    print("WARNING: Revisar asignacion de proporciones, no deben sumar más de 100")

```

```python
#PARTE 3 MISION 7: OBTENER LA CANTIDAD DE BECAS POR NIVEL DEL CRITERIO ELEGIDO

#Cantidad de becas inicial

cantidad_becas_inicial=[]
for i in range (len(proporciones)):
    cantidad_becas_inicial.append(math.floor(proporciones[i]*becas))

#Los elementos de esta lista son la proporción elegida 
cantidad_becas_por_criterio_final=[]
for i in range(len(lista_tuplas_criterio_orden)):
    cantidad_becas_por_criterio_final.append(min(lista_tuplas_criterio_orden[i][1],cantidad_becas_inicial[i]))
    
print(cantidad_becas_por_criterio_final)
residuo_criterio_prop=becas-sum(cantidad_becas_por_criterio_final)
print("Dado el criterio y los porcentajes por nivel elegidos faltaron "+ str(residuo_criterio_prop)+ " becas por asignar.")

```

```python
#PARTE 4 MISION 7: IDENTIFICAR LOS BECADOS POR EL CRITERIO ELEGIDO
def rebanado2(listabecas, listaniveles,indicenivel,listabecados):
    listarebanada=[]
    for i in range(len(listabecas)):
        if listabecas[i][posicion]==listaniveles[indicenivel][0]:
            listarebanada.append(listabecas[i])
            n=int(listabecados[indicenivel])
    return listarebanada[:n]

becados_criterio=[]
for i in range(len(lista_tuplas_criterio_orden)):
    becados_criterio.extend(rebanado2(tuplas_becas_edad,lista_tuplas_criterio_orden,i,cantidad_becas_por_criterio_final))

print("La cantidad de registros becados son "+ str(len(becados_criterio)))
```
